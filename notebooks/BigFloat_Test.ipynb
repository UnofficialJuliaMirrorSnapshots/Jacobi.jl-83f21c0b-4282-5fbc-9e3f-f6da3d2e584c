```julia
using Jacobi
```

```julia
#fun(x,k=1) = cos(k*x) + 1
#ifun(x,k=1) = x + sin(k*x) / k
fun(x) = cos(x)
ifun(x) = -x.*sin(x) + sin(x) - cos(x)
dfun(x) = -sin(x)
```

```julia
function calc_integr(Q, a, b, ::Type{QT}=Jacobi.GLJ, ::Type{T}=Float64) where {T<:Number, QT<:Jacobi.QUADRATURE_TYPE}
    
    z = Jacobi.qzeros(QT, Q, a, b, T)
    w = Jacobi.qweights(QT, z, a, b)
    
    y = fun.(z)
    
    return sum(w .* y)
end

calc_integr(50, 1, 0, Jacobi.GLJ, BigFloat)
```

```julia
setprecision(1024)

Iex(::Type{T}=Float64) where {T<:Number} = ifun(one(T)) - ifun(-one(T))
Iex(BigFloat)
```

```julia
Q = 2:40
tipo =BigFloat
Ie = Iex(tipo)
QT = Jacobi.GRJP
a = 1
b = 0
err = [abs(calc_integr(q, a, b, QT, tipo) - Ie) for q in Q];
```

```julia
using PyPlot
```

```julia
semilogy(Q, err, "o--")
```

```julia
function calc_deriv_err(Q, ::Type{QT}=Jacobi.GLJ, ::Type{T}=Float64) where {T<:Number, QT<:Jacobi.QUADRATURE_TYPE}
    
    z = Jacobi.qzeros(QT, Q, 0, 0, T)
    w = Jacobi.qweights(QT, z, 0, 0)
    D = Jacobi.qdiff(QT, z)
    
    y = fun.(z)
    yex = dfun.(z)
    return maximum(abs, D * y - yex)
end
```

```julia
Q = 2:40
tipo =BigFloat
QT = Jacobi.GLJ
derr = [calc_deriv_err(q, QT, tipo)  for q in Q];
```

```julia
semilogy(Q, derr, "o--")
```
